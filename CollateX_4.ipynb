```python
#Levenshtein distance matrix of two sentences
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import Levenshtein

# Define the two sentences
sentence1 = "the gray fox jumps over the lazy dog"
sentence2 = "the gray fox jumps over the lazy dog"




# Split the sentences into words
words1 = sentence1.split()
words2 = sentence2.split()


# Initialize the distance matrix
distance_matrix = np.zeros((len(words1), len(words2)))

# Calculate the Levenshtein distance for each pair of words
for i, word1 in enumerate(words1):
    for j, word2 in enumerate(words2):
        distance_matrix[i, j] = Levenshtein.distance(word1, word2)

# Plot the heatmap
plt.figure(figsize=(10, 8))
ax = sns.heatmap(distance_matrix, annot=True, fmt=".0f", cmap="YlGnBu", xticklabels=words2, yticklabels=words1)
plt.title("Levenshtein Distance Heatmap")
plt.xlabel("Sentence 2 Words")

plt.ylabel("Sentence 1 Words")

# Adjust the position of the x-axis labels to be above the heatmap
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Rotate the x-axis labels for better readability
#plt.xticks(rotation=45, ha='left')

plt.show()
```
